```python
from IPython import display
categ=da.select_dtypes(include=['int64']).columns
for column in categ:
    da[column].value_counts().sort_values().plot(kind='barh');
    plt.title(f"{column} Distribution")
    plt.show()
    input("Press Enter to see other graphs...")
    display.clear_output()
```
